```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
import numpy as np
plt.style.use('seaborn-dark-palette')

#data = pd.DataFrame.from_csv('truth.csv')
data = pd.DataFrame.from_csv('events/evnt_106.csv')
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111, projection='3d')

xs = np.hstack(data.X)
ys = np.hstack(data.Y)
zs = np.hstack(data.Z)

ax.scatter(xs, ys, zs)

#ax.plot_wireframe(xs, ys, zs)
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111, projection='3d')

mothers = np.hstack(np.array(data['mother']))
uniq_mom = np.unique(mothers)
#values = cm.viridis(np.linspace(0,1,len(uniq_mom)))
values = cm.Paired(np.linspace(0,1,len(uniq_mom)))
col = dict(zip(uniq_mom, values))


for i,mom in zip(xrange(len(uniq_mom)),uniq_mom):
    temp = data[data['mother'] == mom]
    xs = np.hstack(np.array(temp['X']))
    ys = np.hstack(np.array(temp['Y']))
    zs = np.hstack(np.array(temp['Z']))
    ax.scatter(xs, ys, zs,alpha=1,color=col[mom],label=r'%s' %(mom))
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111, projection='3d')

IDs = np.hstack(np.array(data['ID']))
uniq_ID = np.unique(IDs)
#values = cm.viridis(np.linspace(0,1,len(uniq_mom)))
values = cm.Paired(np.linspace(0,1,len(uniq_ID)))
col = dict(zip(uniq_ID, values))


for i,ID in zip(xrange(len(uniq_ID)),uniq_ID):
    temp = data[data['ID'] == ID]
    xs = np.hstack(np.array(temp['X']))
    ys = np.hstack(np.array(temp['Y']))
    zs = np.hstack(np.array(temp['Z']))
    ax.scatter(xs, ys, zs,alpha=1,color=col[ID],label=r'%s' %(ID))
```
